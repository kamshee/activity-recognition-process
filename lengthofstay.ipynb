# Plot univariate distributions
- histogram with fitted kernel density estimate (KDE)

## 1. Length of Stay

```python
# x = np.random.normal(size=100)
sns.distplot(los, bins=10);
plt.title('LOS')
plt.show()
```

```python
# Histogram with rug plot (tick for every observation)
sns.distplot(los, bins=10, kde=False, rug=True)
plt.title('LOS')
plt.show()
```

```python
sns.distplot(los, bins=10,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
plt.title('LOS')
# plt.legend(loc='lower right')
plt.xlabel('Days')
plt.ylabel('Likelihood of occurrence')
plt.show()
```

## 2. Length of data collection

```python
# x = np.random.normal(size=100)
sns.distplot(lodc, bins=10);
plt.title('LODC')
plt.show()
```

```python
# Histogram with rug plot (tick for every observation)
sns.distplot(lodc, bins=10, kde=False, rug=True)
plt.title('LODC')
plt.show()
```

```python
sns.distplot(lodc, bins=10,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
plt.title('LODC')
# plt.legend(loc='lower right')
plt.xlabel('Days')
plt.ylabel('Likelihood of occurrence')
plt.show()
```

# 3. Overlaid CDFs

```python
sns.distplot(los, kde_kws=dict(cumulative=True), label='LOS')
sns.distplot(lodc, kde_kws=dict(cumulative=True), label='LODC')
plt.title('LOS vs LODC CDFs')
plt.xlabel('Days')
plt.ylabel('Likelihood of occurrence')
plt.legend(loc='lower right')
plt.show()
```

# KDE (kernel density estimate)

```python
# overlaid KDEs
sns.kdeplot(los['Length of Stay'].tolist(), bw=4, label='LOS', legend=True)
sns.kdeplot(lodc['Length of Data'].tolist(), bw=4, label='LODC', legend=True)
plt.title('LOS vs LODC kernel density estimate (KDE)')
plt.show()
```
