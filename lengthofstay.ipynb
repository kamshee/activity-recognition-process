```python
# import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# import re
import copy
%matplotlib inline
```

# Real Data

```python
# Length of Stay (LOS)
los = pd.read_csv('Documents/GitHub/LOS.csv', header=None)
lodc = pd.read_csv('Documents/GitHub/LODC.csv', header=None)
```

```python
print('length of los: ',len(los), 'length of lodc: ',len(lodc))
```

# Plot univariate distributions
- histogram with fitted kernel density estimate (KDE)

## 1. Length of Stay

```python
# x = np.random.normal(size=100)
sns.distplot(los, bins=10);
plt.title('LOS')
plt.show()
```

```python
# Histogram with rug plot (tick for every observation)
sns.distplot(los, bins=10, kde=False, rug=True)
plt.title('LOS')
plt.show()
```

```python
sns.distplot(los, bins=10,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
plt.title('LOS')
# plt.legend(loc='lower right')
plt.xlabel('Days')
plt.ylabel('Likelihood of occurrence')
plt.show()
```

## 2. Length of data collection

```python
# x = np.random.normal(size=100)
sns.distplot(lodc, bins=10);
plt.title('LODC')
plt.show()
```

```python
# Histogram with rug plot (tick for every observation)
sns.distplot(lodc, bins=10, kde=False, rug=True)
plt.title('LODC')
plt.show()
```

```python
sns.distplot(lodc, bins=10,
             hist_kws=dict(cumulative=True),
             kde_kws=dict(cumulative=True))
plt.title('LODC')
# plt.legend(loc='lower right')
plt.xlabel('Days')
plt.ylabel('Likelihood of occurrence')
plt.show()
```

# KDE

```python
sns.kdeplot(los.iloc[:,0].values, label='default bandwidth', legend=True)
sns.kdeplot(los.iloc[:,0].values, bw=3, label="bandwidth: 3")
plt.show()
```
