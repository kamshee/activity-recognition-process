# # Visualize signal processing 
- gradient vs derivative using diff()

```python
# load data
x = clip_data[0]['accel']['data'][0]
# Take subjset of data: using accelerometer x-axis
tempx = x.iloc[:10,0]
# derivative = difference/32
derivative = np.gradient(tempx, 32)
# deriv from diff
testx = np.diff(tempx,axis=0)/32

fig, ax = plt.subplots(3, sharex=True, figsize=(6, 6))

ax[0].plot(x.index[:10], tempx)
ax[1].plot(x.index[:10], derivative, color="r")
ax[2].plot(x.index[:9], testx, color="g")
```
